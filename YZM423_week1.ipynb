# Phi-3


*   AutoModelForCausalLM: Dil modelleri için otomatik model yükleme sınıfı
*   AutoTokenizer: Metni model için uygun formata dönüştüren tokenizer'ı otomatik yükleyen sınıf
*   "Auto" sınıfları, model adına göre otomatik olarak doğru model ve tokenizer türünü seçer

"microsoft/Phi-3-mini-4k-instruct": Hugging Face Hub'dan yüklenecek model adı

Microsoft tarafından geliştirilen Phi-3 ailesinin küçük versiyonu
4k token bağlam uzunluğu destekler
"instruct" versiyonu, talimat takip etmek için fine-tune edilmiş


device_map="cuda": Modeli GPU'ya (CUDA) yükler

Eğer GPU yoksa hata verir
Alternatifler: "cpu", "auto" (otomatik cihaz seçimi)


torch_dtype="auto": Veri tipini otomatik seçer

Model konfigürasyonuna göre en uygun veri tipini (float16, bfloat16, float32) seçer
Bellek kullanımını optimize eder


trust_remote_code=False: Güvenlik ayarı

Model ile birlikte gelen custom kod çalıştırılmasına izin vermez
Güvenlik için önerilen ayar




```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# Load model and tokenizer
model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",
    device_map="cuda",
    torch_dtype="auto",
    trust_remote_code=False,
)
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")
```

"text-generation"

Pipeline türünü belirtir
Metin üretimi (text generation) görevi için yapılandırır
Diğer seçenekler: "text-classification", "question-answering", "summarization" vb.

model=model

Önceden yüklenen Phi-3 modelini kullanır
Pipeline'ın hangi dil modeliyle çalışacağını belirtir

tokenizer=tokenizer

Önceden yüklenen tokenizer'ı kullanır
Giriş metnini token'lara dönüştürme ve çıktıyı metne çevirme işlemlerinde kullanılır


do_sample=False

Deterministik üretim modunu etkinleştirir
False: Her seferinde aynı giriş için aynı çıktıyı üretir (greedy decoding)
True olsaydı: Rastgelelik ekleyerek her seferinde farklı çıktılar üretirdi
Test ve tutarlı sonuçlar için ideal,

return_full_text=False

Önemli ayar: Sadece yeni üretilen metni döndürür
True olsaydı: "Merhaba" + " dünya nasılsın?" şeklinde tam metin dönerdi
False olduğunda: Sadece " dünya nasılsın?" kısmını döndürür
Genellikle sohbet uygulamaları için tercih edilir



```python
from transformers import pipeline

# Create a pipeline
generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    return_full_text=False,
    max_new_tokens=500,
    do_sample=False
)
```

messages: Chat formatında mesajları tutan liste
{"role": "user", "content": "..."}: Sohbet mesajının yapısı

"role": "user": Bu mesajın kullanıcıdan geldiğini belirtir
"content": "Create a funny joke about chickens.": Gerçek mesaj içeriği

*   generator: Önceden tanımlanmış pipeline nesnesini çağırır
*    messages: Chat formatındaki giriş verisi
*    output: Pipeline'ın döndürdüğü sonuç

*   output[0]: Pipeline sonucu bir liste döndürür, ilk (ve genelde tek) sonucu alır
*   ["generated_text"]: Üretilen metnin anahtarı

```python
# The prompt (user input / query)
messages = [
    {"role": "user", "content": "Create a funny joke about chickens."}
]

# Generate output
output = generator(messages)
print(output[0]["generated_text"])
```
